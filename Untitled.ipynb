```python
import ast

def convert(text):
    if isinstance(text, list):
        return text
    try:
        L = []
        for i in ast.literal_eval(text):
            L.append(i['name'])
        return L
    except:
        return []
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load MovieLens ratings
ratings = pd.read_csv("u.data", sep="\t", names=["user_id", "movie_id", "rating", "timestamp"])

# Load MovieLens movie titles
movies = pd.read_csv("u.item", sep="|", encoding="latin-1", header=None, usecols=[0, 1])
movies.columns = ["movie_id", "title"]
movies["clean_title"] = movies["title"].str.replace(r"\s\(\d{4}\)", "", regex=True).str.lower()

# Load TMDB data
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb["clean_title"] = tmdb["title"].str.lower()

# Merge datasets
merged_movies = movies.merge(tmdb, on="clean_title", how="inner")
data = ratings.merge(merged_movies, on="movie_id")

# Create binary target variable
data["liked"] = data["rating"].apply(lambda x: 1 if x >= 4 else 0)

# Select features
features = data[["popularity", "vote_average", "vote_count"]]
target = data["liked"]

# Fill missing values
features = features.fillna(0)

# Scale features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# Train Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
def recommend_movies(model, scaler, movie_data, top_n=10):
    # Filter only required features
    features = movie_data[["popularity", "vote_average", "vote_count"]].fillna(0)
    
    # Scale the features
    features_scaled = scaler.transform(features)
    
    # Predict whether the movie is liked (1) or not (0)
    predictions = model.predict(features_scaled)
    
    # Add predictions back to the movie data
    movie_data = movie_data.copy()
    movie_data["predicted_like"] = predictions
    
    # Filter only liked movies and return top_n by popularity
    recommended = movie_data[movie_data["predicted_like"] == 1]
    recommended = recommended.sort_values(by="popularity", ascending=False).head(top_n)
    
    return recommended[["original_title", "popularity", "vote_average"]]
```

```python
recommended_movies = recommend_movies(model, scaler, merged_movies)
print("Top Recommended Movies Based on Logistic Regression:")
print(recommended_movies)
```
